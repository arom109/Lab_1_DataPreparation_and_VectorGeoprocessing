```python
#using this list, separate out soil layers and tables into two lists

soils = ['soilmu_a_co001', 'soilmu_a_co618', 'soilmu_a_co641', 'soilmu_a_co642', 'soilmu_a_co643', 'soilmu_a_co644', 'soilmu_a_co645', 'soilmu_a_co651', 'soilmu_a_co653']
tables = ['muaggatt_co001', 'muaggatt_co618', 'muaggatt_co641', 'muaggatt_co642', 'muaggatt_co643', 'muaggatt_co644', 'muaggatt_co645', 'muaggatt_co651', 'muaggatt_co653']
```

```python
#now merge the lists using nested for loops

merge_list = []

for i in soils:
    for j in tables:
        if i[-5:] == j[-5:]:
            soil_df = gpd.read_file(r'C:\Users\arom\Documents\School\Grad School\Fall 2020\Automation\week1\Lab01\lab1\lab1.gpkg', layer = i)
            table_df = gpd.read_file(r'C:\Users\arom\Documents\School\Grad School\Fall 2020\Automation\week1\Lab01\lab1\lab1.gpkg', layer = j)
            table_df= table_df.drop(columns='geometry')
            merge1 = pd.merge(soil_df, table_df, left_on='MUSYM', right_on='musym')
            merge_list.append(merge1)
```

```python
merge_list
```

```python
# add the new mapunit field, populate, merge all 9 feature classes into 1

merge1['mapunit'] = i[-5:]
new_class = pd.concat(merge_list)
```

```python
new_class
```

```python
# open watershed data, use sjoin to join watersheds and new_class

watershed_df = gpd.read_file(r'C:\Users\arom\Documents\School\Grad School\Fall 2020\Automation\week1\Lab01\lab1\lab1.gpkg', layer = 'wbdhu8_lab1')

new_class_watershed = gpd.sjoin(new_class, watershed_df,  how='left', op="intersects")
```
